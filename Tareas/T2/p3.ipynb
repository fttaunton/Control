```python
s = c.tf('s')
plt.figure()
valores_alpha = [1, 2, 4, 8, 10, 16, 32, 50, 64, 100]

a = valores_alpha[0]
den = (5*s/a + 1) * (s**2 + 10*s + 50)
syst = 50/den
H = c.tf(syst)
t, y = c.step_response(H)
plt.plot(t,y)
plt.savefig("graf" + str(a) + ".png")
```

```python
c.pzmap(H, plot=True)
plt.savefig("polos" + str(a) + ".png")
```
